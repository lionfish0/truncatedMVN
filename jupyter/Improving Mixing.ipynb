```python
from truncatedMVN import TruncMVN,TruncMVNreparam
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
mean = np.array([0.5,0.5])
cov = np.array([[0.3,-0.29999],[-0.29999,0.3]])
planes = np.array([[1,0],[-1,1],[-2,1]])
tm = TruncMVN(mean,cov,planes,burnin=10,thinning=1)
plt.figure()
tm.plot()
plt.xlim([-0.1,0.5])
plt.ylim([0.5,1])
print(tm.compute_gelman_rubin())


tm = TruncMVNreparam(mean,cov,planes,burnin=10,thinning=1)
print(tm.compute_gelman_rubin())
plt.figure()
tm.plot()
plt.xlim([-0.1,0.5])
plt.ylim([0.5,1])
```
